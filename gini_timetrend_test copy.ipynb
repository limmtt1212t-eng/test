# Gini Time Trend Test

Notebook для тестирования функций из `gini_timetrend.py`.

Используем файл `gini_timetrend_sample.xlsx` (лист `data`).

```python
import pandas as pd
import matplotlib.pyplot as plt

from gini_timetrend import plot_gini_dynamics_binary

%matplotlib inline
```

```python
df = pd.read_excel('gini_timetrend_sample.xlsx', sheet_name='data')
df.head()
```

```python
fig, ax = plot_gini_dynamics_binary(
    df1=df,
    observed_col='y_true',
    predicted_col='y_pred',
    date_col='date',
    freq='M',
    min_observations=100,
    alpha_95=0.05,
    alpha_99=0.01,
    figsize=(14, 8),
    title='Динамика Gini по месяцам (пример)'
)
plt.show()
```

```python
fig, ax = plot_gini_dynamics_binary(
    df1=df,
    observed_col='y_true',
    predicted_col='y_pred',
    date_col='date',
    freq='Q',   # или 'Y'
)
plt.show()
```
